```python
gen_path = "/root/Auditing_Test_for_LMs/"
model_output_path = os.path.join(gen_path, "model_outputs")
model_score_path = os.path.join(gen_path, "model_scores")

with open(os.path.join(gen_path, "high_toxicity_indices.json")) as f:
    high_toxicity_indices = json.load(f)
```

```python
# Look at particularly toxic ones

model_name = "sentence_perturbation-Meta-Llama-3-8B-Instruct"
# model_name = "LLama-3-8b-Uncensored"
seed = "seed1000"

cont_path = os.path.join(model_output_path, f"{model_name}_{seed}/{model_name}_continuations_{seed}.json")
score_path = os.path.join(model_score_path, f"{model_name}_{seed}/perspective_continuation_scores.json")
full_score_path = os.path.join(model_score_path, f"{model_name}_{seed}/perspective_scores.json")

with open(cont_path) as f:
    conts = json.load(f)

with open(score_path) as f:
    scores = json.load(f)
    
with open(full_score_path) as f:
    full_scores = json.load(f)

prompts = conts["prompts"]
conts = conts["continuations"]
scores = scores["perspective_scores"]
full_scores = full_scores["perspective_scores"]

hightoxprompts = [prompts[i] for i in high_toxicity_indices]
hightoxconts = [conts[i] for i in high_toxicity_indices]
hightoxscores = [scores[i] for i in high_toxicity_indices]
fulltoxscores = [full_scores[i] for i in high_toxicity_indices]
```

```python
def sort_list_by_other(list_to_sort, ordering_list):
    return [x for _, x in sorted(zip(ordering_list, list_to_sort))]

indices = list(range(len(prompts)))
sorted_indices = sort_list_by_other(indices, scores)

diff_cont_full = [scores[i] - full_scores[i] for i in range(len(scores))]

sorted_by_diff_indices = sort_list_by_other(indices, diff_cont_full)
```
